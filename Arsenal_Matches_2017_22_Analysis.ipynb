```python
#replace home with 0 and away with 1
df['HoAw'] = df['HoAw'].replace({'home':0,'away':1})
```

```python
df.head()
```

```python
df.columns
```

```python
#finding the outcomes of the matches
df['ScoreDifference'] = df['ArsenalScore'] - df['OpponentScore']
df['Outcome'] = df['ScoreDifference'].apply(lambda x: 'win' if x>0 else 'loss' if x < 0 else 'tie')
```

```python
df.head()
```

```python
#points earned per game
df['PointsEarned'] = df['ScoreDifference'].apply(lambda x: 3 if x > 0 else 1 if x == 0 else 0)
```

```python
#now let's analyze by the specific seasons, i.e., 2017/18, etc.
year_17_18 = df[df['Season'] == '2017/18']["Outcome"].value_counts()
wins_17_18 = year_17_18[0]
loss_17_18 = year_17_18[1]
ties_17_18 = year_17_18[2]

year_18_19 = df[df['Season'] == '2018/19']["Outcome"].value_counts()
wins_18_19 = year_18_19[0]
loss_18_19 = year_18_19[1]
ties_18_19 = year_18_19[2]

year_19_20 = df[df['Season'] == '2019/20']["Outcome"].value_counts()
wins_19_20 = year_19_20[0]
loss_19_20 = year_19_20[1]
ties_19_20 = year_19_20[2]

year_20_21 = df[df['Season'] == '2020/21']["Outcome"].value_counts()
wins_20_21 = year_20_21[0]
loss_20_21 = year_20_21[1]
ties_20_21 = year_20_21[2]

year_21_22 = df[df['Season'] == '2021/22']["Outcome"].value_counts()
wins_21_22 = year_21_22[0]
loss_21_22 = year_21_22[1]
ties_21_22 = year_21_22[2]

year_22_23 = df[df['Season'] == '2022/23']["Outcome"].value_counts()
wins_22_23 = year_22_23[0]
loss_22_23 = year_22_23[1]
ties_22_23 = year_22_23[2]
```

```python
#pie plots showing distribution
plt.pie([wins_17_18,loss_17_18,ties_17_18],labels=["Wins","Losses","Ties"],autopct="%.2f")
plt.title("17/18 Performance")
plt.show()

plt.pie([wins_18_19,loss_18_19,ties_18_19],labels=["Wins","Losses","Ties"],autopct="%.2f")
plt.title("18/19 Performance")
plt.show()

plt.pie([wins_19_20,loss_19_20,ties_19_20],labels=["Wins","Losses","Ties"],autopct="%.2f")
plt.title("19/20 Performance")
plt.show()

plt.pie([wins_20_21,loss_20_21,ties_20_21],labels=["Wins","Losses","Ties"],autopct="%.2f")
plt.title("20/21 Performance")
plt.show()

plt.pie([wins_21_22,loss_21_22,ties_21_22],labels=["Wins","Losses","Ties"],autopct="%.2f")
plt.title("21/22 Performance")
plt.show()

plt.pie([wins_22_23,loss_22_23,ties_22_23],labels=["Wins","Losses","Ties"],autopct="%.2f")
plt.title("22/23 Performance")
plt.show()
```

```python
num_17_18 = df[df['Season'] == "2017/18"].value_counts()
n1 = len(num_17_18)
l1 = []
l2 = []
for i in range(n1):
  l1.append(df['Tour'][i])
  l2.append(df["PointsEarned"][i])

plt.figure(figsize=(10,3))
plt.plot(l1, l2, color="b", marker="o", markersize=5)
plt.xlabel("MatchWeek")
plt.ylabel("Points earned")
plt.title("2017/18 Season")
plt.show()
print()

num_18_19 = df[df['Season'] == "2018/19"].value_counts()
n2 = n1 * 2
l1 = []
l2 = []
for i in range(n1,n2):
  l1.append(df['Tour'][i])
  l2.append(df["PointsEarned"][i])

plt.figure(figsize=(10,3))
plt.plot(l1, l2, color="b", marker="o", markersize=5)
plt.xlabel("MatchWeek")
plt.ylabel("Points earned")
plt.title("2018/19 Season")
plt.show()
print()

num_19_20 = df[df['Season'] == "2019/20"].value_counts()
n3 = n1 * 3
l1 = []
l2 = []
for i in range(n2,n3):
  l1.append(df['Tour'][i])
  l2.append(df["PointsEarned"][i])

plt.figure(figsize=(10,3))
plt.plot(l1, l2, color="b", marker="o", markersize=5)
plt.xlabel("MatchWeek")
plt.ylabel("Points earned")
plt.title("2019/20 Season")
plt.show()
print()

num_20_21 = df[df['Season'] == "2020/21"].value_counts()
n4 = n1 * 4
l1 = []
l2 = []
for i in range(n3,n4):
  l1.append(df['Tour'][i])
  l2.append(df["PointsEarned"][i])

plt.figure(figsize=(10,3))
plt.plot(l1, l2, color="b", marker="o", markersize=5)
plt.xlabel("MatchWeek")
plt.ylabel("Points earned")
plt.title("2020/21 Season")
plt.show()
print()

num_21_22 = df[df['Season'] == "2021/22"].value_counts()
n5 = n1 * 5
l1 = []
l2 = []
for i in range(n4,n5):
  l1.append(df['Tour'][i])
  l2.append(df["PointsEarned"][i])

plt.figure(figsize=(10,3))
plt.plot(l1, l2, color="b", marker="o", markersize=5)
plt.xlabel("MatchWeek")
plt.ylabel("Points earned")
plt.title("2021/22 Season")
plt.show()
print()

num_22_23 = df[df['Season'] == "2022/23"].value_counts()
n6 = n5 + len(num_22_23)
l1 = []
l2 = []
for i in range(n5,n6):
  l1.append(df['Tour'][i])
  l2.append(df["PointsEarned"][i])

plt.figure(figsize=(12,3))
plt.plot(l1, l2, color="b", marker="o", markersize=5)
plt.xlabel("MatchWeek")
plt.ylabel("Points earned")
plt.title("2022/23 Season")
plt.show()
```

```python
goals_scored_17_18 = 0
goals_conceded_17_18 = 0
for i in range(n1):
  goals_scored_17_18 += df['ArsenalScore'][i]
  goals_conceded_17_18 += df['OpponentScore'][i]

goals_scored_18_19 = 0
goals_conceded_18_19 = 0
for i in range(n1,n1*2):
  goals_scored_18_19 += df['ArsenalScore'][i]
  goals_conceded_18_19 += df['OpponentScore'][i]

goals_scored_19_20 = 0
goals_conceded_19_20 = 0
for i in range(n1*2,n1*3):
  goals_scored_19_20 += df['ArsenalScore'][i]
  goals_conceded_19_20 += df['OpponentScore'][i]

goals_scored_20_21 = 0
goals_conceded_20_21 = 0
for i in range(n1*3,n1*4):
  goals_scored_20_21 += df['ArsenalScore'][i]
  goals_conceded_20_21 += df['OpponentScore'][i]

goals_scored_21_22 = 0
goals_conceded_21_22 = 0
for i in range(n1*4,n1*5):
  goals_scored_21_22 += df['ArsenalScore'][i]
  goals_conceded_21_22 += df['OpponentScore'][i]

l1 = [goals_scored_17_18, goals_scored_18_19, goals_scored_19_20, goals_scored_20_21, goals_scored_21_22]
l2 = [goals_conceded_17_18, goals_conceded_18_19, goals_conceded_19_20, goals_conceded_20_21, goals_conceded_21_22]

plt.figure(figsize=(12,5))
plt.barh(["2017/18","2018/19","2019/20","2020/21","2021/22"],l1,color="yellow",label="Goals Scored")
plt.barh(["2017/18","2018/19","2019/20","2020/21","2021/22"],l2,color="red", label="Goals Conceded")
plt.legend()
plt.ylabel("Season")
plt.xlabel("Number of Goals")
plt.show()
```
